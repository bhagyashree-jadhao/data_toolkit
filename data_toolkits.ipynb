### 14) Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and Perform the following tasks using Matplotlib: a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and right y-axis for 'Humidity').

b) Label the x-axis as 'Date'.

c) Set the title of the plot as 'Temperature and Humidity Over Time'.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create the time-series dataset
data = {
    'Date': pd.date_range(start='2024-07-01', periods=10, freq='D'),
    'Temperature': [30, 32, 31, 29, 28, 27, 26, 25, 24, 23],
    'Humidity': [70, 65, 68, 72, 75, 80, 85, 78, 77, 76]
}

df = pd.DataFrame(data)
print(df)

# Extract only the day part from the Date column
df['Day'] = df['Date'].dt.day

# Plot the data with different y-axes
fig, ax1 = plt.subplots()

# Plot Temperature on the left y-axis
ax1.plot(df['Day'], df['Temperature'],'-b', label='Temperature')
ax1.set_xlabel('Day')
ax1.set_ylabel('Temperature (°C)', color='r')
ax1.tick_params(axis='y', labelcolor='r')

# Create a second y-axis for Humidity
ax2 = ax1.twinx()
ax2.plot(df['Day'], df['Humidity'], label='Humidity')
ax2.set_ylabel('Humidity (%)', color='b')
ax2.tick_params(axis='y', labelcolor='b')

# Set the title of the plot
plt.title('Temperature and Humidity Over Time')

# Display the plot
fig.tight_layout()  # To prevent the labels from overlapping
plt.show()
```

```python
import numpy as np

# Create an array with 1000 samples from a normal distribution
data = np.random.normal(loc=0, scale=1, size=1000)


print(data)
```

```python
from scipy.stats import norm
# Plot a histogram with 30 bins
plt.hist(data, bins=30, density=True, edgecolor='black', alpha=0.6, color='skyblue')

# Calculate the PDF
xmin, xmax = plt.xlim()  # Get the range of x-axis values from the histogram
x = np.linspace(xmin, xmax, 100)  # Generate x values
p = norm.pdf(x, loc=0, scale=1)  # Calculate the PDF for each x value

# Overlay the PDF on the histogram
plt.plot(x, p, 'r', linewidth=2)

#c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'.
plt.xlabel('Frequency')
plt.ylabel('Probability')

# d) Set the title of the plot as 'Histogram with PDF Overlay'
plt.title('Histogram with PDF Overlay')

plt.show()
```

###  17) Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Generate two random arrays
x = np.random.randn(100)
y = np.random.randn(100)

# Step 2: Determine the quadrant for each point
def get_quadrant(x, y):
    if x >= 0 and y >= 0:
        return 'Quadrant I'
    elif x < 0 and y >= 0:
        return 'Quadrant II'
    elif x < 0 and y < 0:
        return 'Quadrant III'
    else:
        return 'Quadrant IV'

# Apply the quadrant function to each point
quadrants = [get_quadrant(x_val, y_val) for x_val, y_val in zip(x, y)]

# Create a DataFrame for easier plotting
df = pd.DataFrame({'x': x, 'y': y, 'Quadrant': quadrants})

# Step 3: Create the scatter plot with Seaborn
sns.scatterplot(data=df, x='x', y='y', hue='Quadrant', palette='bright', style='Quadrant')

# Label the axes
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# Set the title
plt.title('Quadrant-wise Scatter Plot')

# Display the legend
plt.legend(title='Quadrants')


plt.show()
```

### 18) *58 With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine Wave Function'8

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Output the plot directly in the notebook (or use output_file for saving it)
output_notebook()

# Step 1: Generate the sine wave data
x = np.linspace(0, 4 * np.pi, 500)
y = np.sin(x)

# Step 2: Create a figure
p = figure(title="Sine Wave Function", x_axis_label='X Axis', y_axis_label='Y Axis')

# Step 3: Plot the sine wave
p.line(x, y, legend_label="Sine Wave", line_width=2)

# Step 4: Add grid lines (enabled by default)
p.grid.grid_line_alpha = 0.6  # Adjust the transparency of the grid lines

# Step 5: Set the title and labels
p.title.text = "Sine Wave Function"
p.xaxis.axis_label = "X Axis"
p.yaxis.axis_label = "Y Axis"

# Step 6: Show the plot
show(p)
```

### 20) Using Plotly, create a basic line plot of a randomly generated dataset, label the axes, and set the title as 'Simple Line Plot'.

```python
import plotly.graph_objects as go
import numpy as np

# Generate a random dataset
x = np.linspace(0, 10, 100)
y = np.random.randn(100)

# Create the line plot
fig = go.Figure()

# Add the line trace
fig.add_trace(go.Scatter(x=x, y=y, mode='lines', name='Random Data'))

# Add labels and title
fig.update_layout(
    title='Simple Line Plot',
    xaxis_title='X-axis',
    yaxis_title='Y-axis'
)

# Display the plot
fig.show()
```

### 21) Using Plotly, create an interactive pie chart of randomly generated data, add labels and percentages, set the title as 'Interactive Pie Chart'.

```python

import plotly.graph_objects as go
import numpy as np

# Generate random data
np.random.seed(0)
values = np.random.randint(10, 100, 5)
labels = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']

# Create a pie chart
fig = go.Figure(data=[go.Pie(labels=labels, values=values, hole=.3,
                             hoverinfo='label+percent', textinfo='percent')])

# Add title
fig.update_layout(title_text='Interactive Pie Chart')

# Display the pie chart
fig.show()
```
